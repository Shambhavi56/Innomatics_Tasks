## 1. Smart Light Controller

```python
class Smart_Light_Controller:
    def __init__(self,Light_Name,Action):
      self.Light_Name = Light_Name
      self.Action =Action
    def control_light(self):
        if self.Action == "ON":
          return f"{self.Light_Name} is {self.Action}"
        else:
          return f"{self.Light_Name} is {(self.Action).upper()}"

Light_Name = input( "Light Name: ")
Action = input("Action: ")
light = Smart_Light_Controller(Light_Name, Action)

print(light.control_light())
```

## 3. Mobile Contact Record

```python
class Mobile_Contact_Record:
  def __init__(self, name, country_code , phone_number):
    self.name = name
    self.country_code = country_code
    self.phone_number = phone_number
  def display_contact(self):
    print(f"Name: {self.name}\nPhone Number:+{self.country_code}{self.phone_number}")

# get username and must be alphanumeric
while True:
 name_input= input("Name: ")
 if all(c.isalnum() or c.isspace() for c in name_input):
  Name = name_input
 break
else:
  print("Invalid input. Please enter alphanumeric characters only for the name.")

# restrict the input to digit
while True:
    code_input = input("Country Code: ")
    if code_input.isdigit() and len(code_input)==2:
        country_code = int(code_input)
        break
    print("Invalid input. Please enter digits only for the country code.")

# restrict the input to 10 digits only
while True:
   phone_number = input("Phone Number:")
   if phone_number.isdigit() and len(phone_number) == 10:
     break
   else:
    print("Invalid input. Please enter a 10-digit phone number.")

contact = Mobile_Contact_Record(Name, country_code, phone_number)
contact.display_contact()
```

##4. Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        # Using Unicode \u20B9 for the Rupee symbol (â‚¹)
        print(f"Product: {self.name}")
        print(f"Price: \u20B9{self.price}")

# Input section
prod_name = input("Product Name: ")

# Validation to ensure price is a number
while True:
    price_input = input("Price: ")
    if price_input.isdigit():
        price = int(price_input)
        break
    print("Invalid price. Please enter digits only.")

# Generate the tag
tag = Product(prod_name, price)
tag.display_tag()
```

##5. Movie Rating Display System

Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
Streaming platforms
ðŸ“¥ Sample Input
Movie Name: Inception
Rating: 4.8
ðŸ“¤ Expected Output
Movie: Inception
Rating: 4.8 / 5


```python
class Movie_Rating:
    def __init__(self, Name, Rating):
        self.Name = Name
        self.Rating = Rating
    def display_movie(self):
        # Result is formatted to show 1 decimal place (e.g., 4.5/5)
        print(f"Movie: {self.Name}\nRating: {self.Rating:.1f}/5")

# Outside the class
Movie_Name = input("Movie Name: ")
Rating_Value = float(input("Rating: "))

movie = Movie_Rating(Movie_Name, Rating_Value)
movie.display_movie()

```
