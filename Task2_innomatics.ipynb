## Problem 1: Unique Words in a Sentence


```python

class UniqueWords:
    def find_unique(self, sentence):
        words = sentence.split()          # Split sentence into words
        unique_words = set(words)         # Remove duplicates
        print("Unique words count:", len(unique_words))
        print("Unique words:", unique_words)


# Taking user input
sentence = input("Enter sentence: ")

obj = UniqueWords()
obj.find_unique(sentence)
```

## Problem 2: Highest Salary from Employee Data


```python
# Taking dictionary input from user

n = int(input("Enter number of employees: "))
employees = {}

for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

# Initialize using first employee
names = list(employees.keys())
highest_employee = names[0]
highest_salary = employees[highest_employee]

# Compare salaries
for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_employee = name

# Exact expected output format
print("Highest Salary:", highest_employee, "-", highest_salary)


```

## Find Maximum and Minimum Values


```python
# Taking list input from user (comma separated)
numbers = input("Enter numbers separated by comma: ").split(",")

# Converting each value to integer
numbers = [int(num) for num in numbers]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Compare each element
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)

```

## 4. Count Products Above a Price Threshold


```python
# Taking input from user
prices = input("Enter product prices separated by comma: ")

# Convert input string to list of integers
prices = prices.replace(",", " ").split()
prices = [int(price) for price in prices]

count = 0

# Compare each price with 1000
for price in prices:
    if price > 1000:
        count += 1

# Exact expected output format
print("Products above 1000:", count)
```

## 5: Calculate Attendance Percentage

```python
# Taking input from user
attendance = input("Enter attendance (P/A separated by comma): ")

# Convert input into list
attendance = attendance.replace(",", " ").split()

present_count = 0

# Count number of "P"
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

# Exact expected output format
print("Attendance Percentage:", percentage)
```

## 6: Remove Duplicate Phone Numbers


```python

# Taking input from user
phone_numbers = input("Enter phone numbers separated by comma: ")

# Convert input into list of integers
phone_numbers = phone_numbers.replace(",", " ").split()
phone_numbers = [int(number) for number in phone_numbers]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Exact expected output format
print("Unique phone numbers:", unique_numbers)
```

## 7: Count Character Frequency

```python
# Taking input from user
text = input("Enter a string: ")

frequency = {}

# Counting each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Exact expected output format
print(frequency)
```

## 9: Check if a Key Exists in a Dictionary



```python
# Taking dictionary input from user
n = int(input("Enter number of employees: "))
employees = {}

for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

# Key to check
key_to_check = input("Enter employee name to check: ")

# Using 'in' keyword
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

## 10: Calculate Average Marks

```python
# Taking input from user
marks = input("Enter marks separated by comma: ")

# Convert input into list of integers
marks = marks.replace(",", " ").split()
marks = [int(mark) for mark in marks]

total = 0

# Calculate sum manually
for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

# Exact expected output format
print("Average Marks:", average)
```
