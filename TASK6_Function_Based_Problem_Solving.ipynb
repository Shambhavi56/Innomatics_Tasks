## 1. Smart Parking Lot Management System


```python
def smart_parking_system(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            # Prevent negative count
            if current_parked > 0:
                current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Capacity Check
    if current_parked > capacity:
        status = "Parking Full - Capacity Exceeded!"
    elif current_parked == capacity:
        status = "Parking Full"
    else:
        status = "Available"

    return current_parked, peak_usage, status


# -----------------------------
# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function Call
current, peak, parking_status = smart_parking_system(capacity, vehicle_logs)

# Output
print("Currently Parked Vehicles:", current)
print("Peak Parking Usage:", peak)
print("Parking Status:", parking_status)
```

## 2. Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base calculation
    eta = distance * 4   # 4 minutes per km

    # Traffic adjustment
    if traffic == "Low":
        eta += 0
    elif traffic == "Medium":
        eta += 5
    elif traffic == "High":
        eta += 15

    # Weather adjustment
    if weather == "Clear":
        eta += 0
    elif weather == "Rainy":
        eta += 8
    elif weather == "Storm":
        eta += 15

    return eta


# -----------------------
# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function Call
final_eta = delivery_time_estimator(distance, traffic, weather)

# Output
print("Estimated Delivery Time:", final_eta, "minutes")
```

## 3. Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    # Count booked seats using loop
    booked_count = 0
    for seat in booked_seats:
        booked_count += 1

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open Additional Show"
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider Opening Additional Show"
    else:
        status = "Seats Available"
        suggestion = "No Extra Show Needed"

    return occupancy, status, suggestion


# ---------------------------------
# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Function Call
occupancy_percent, show_status, suggestion = seat_occupancy_analyzer(total_seats, booked_seats)

# Output
print("Occupancy:", occupancy_percent, "%")
print("Show Status:", show_status)
print("Suggestion:", suggestion)
```

## 4. Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):

    total = 0

    # Loop to calculate total CPU usage
    for reading in cpu_readings:
        total += reading

    # Calculate average CPU load
    average_cpu = total / len(cpu_readings)

    # Classification logic
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_cpu, status


# -------------------------
# Sample Input
cpu_data = [45, 60, 70, 85, 90]

# Function Call
avg_cpu, server_status = classify_server_load(cpu_data)

# Output
print("Average CPU Load:", avg_cpu, "%")
print("Server Status:", server_status)
```

## 5. Smart Classroom Resource Usage Monitor


```python
def classroom_resource_monitor(resource_usage):

    overused_resources = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:   # Overuse threshold
            overused_resources.append(resource)

    # Energy alert check
    if len(overused_resources) > 0:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    return overused_resources, energy_alert


# -------------------------
# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function Call
overused, alert = classroom_resource_monitor(usage_data)

# Output Formatting
if overused:
    print("Overused Resources:", ", ".join(overused))
else:
    print("Overused Resources: None")

print("Energy Alert:", alert)
```

## 6. Online Event Registration Capacity Controller


```python
def event_registration_controller(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    # Loop to simulate registration handling
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status


# -------------------------
# Sample Input
event_capacity = 100
registrations = 105

# Function Call
confirmed_users, waitlisted_users, registration_status = event_registration_controller(
    event_capacity, registrations
)

# Output
print("Confirmed Registrations:", confirmed_users)
print("Waitlisted Users:", waitlisted_users)
print("Registration Status:", registration_status)
```
