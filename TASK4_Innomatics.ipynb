## 1: Mobile Recharge Validation System


```python
def recharge_validation():
    valid_plans = [199, 299, 399, 599]

    while True:
        recharge_amount = int(input("Enter the recharge amount: "))

        if recharge_amount < 50:
            print("Amount must be at least ₹50. Try again.")
        elif recharge_amount in valid_plans:
            print(f"Valid recharge amount: ₹{recharge_amount}")
            break
        else:
            print("Invalid plan. Choose from 199, 299, 399, 599.")

recharge_validation()

```

## 2: Inventory Reorder Alert System


```python
product_stocks={}
# accept user input for how many products are to be counted.
# collect the stock for each product
# check stocks for the entire product catalogue
# record alert for products if stock < 15  print "recorder alert" else "stock ok"
def inventory_reorder_alert():
    num_products = int(input("Enter the number of products: "))
    for i in range(num_products):
        product_name = input(f"Enter the name of product {i+1}: ")
        stock = int(input(f"Enter the stock for {product_name}: "))
        product_stocks[product_name] = stock
    for product_name, stock in product_stocks.items():
        if stock <15:
            print(f"Recorder alert: {product_name} stock is {stock}")
        else:
            print(f"stock ok ")
inventory_reorder_alert()


```

## 3: Student Result Processing System


```python
def result_processing():
    while True:
        student_marks = []
        num_subjects = int(input("Enter the number of subjects: "))

        for i in range(num_subjects):
            student_marks.append(int(input(f"Enter marks for subject {i+1}: ")))

        average_marks = sum(student_marks) / num_subjects

        if average_marks >= 50:
            result = "PASS"
        else:
            result = "FAIL"

        print("Average:", average_marks)
        print("Result:", result)

        retry = input("Do you want to process another student? (yes/no): ").lower()
        if retry != "yes":
            break

result_processing()
```

##  4: Cab Fare Estimator with Retry Option

```python
def cab_fare_estimator():
    base_fare = 50
    price_per_km = 12

    while True:
        distance = float(input("Enter distance in kilometers: "))
        peak_hour = input("Is it peak hour? (yes/no): ").lower()

        fare = base_fare + (price_per_km * distance)

        if peak_hour == "yes":
            fare += 0.25 * fare   # 25% extra

        print(f"Total Fare: ₹{fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

cab_fare_estimator()
```

## 5: Employee Attendance Eligibility

```python
def attendance_checker(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_checker(["P","A","P","P"])
```

## 6: Password Strength Checker


```python
def password_checker(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password_checker("password@123")
```
