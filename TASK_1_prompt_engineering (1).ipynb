## Task 1: Replace Hardcoded Prompts

```python
# Task 1: Convert hardcoded prompt into reusable PromptTemplate

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Test
output_1 = template_1.format(topic="Artificial Intelligence")
print("Task 1 Output:\n", output_1)
```

## Task 2: Multi-Input Prompt System

```python
# Task 2: Multi-variable prompt

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"},
]

print("Task 2 Outputs:\n")
for case in test_cases:
    print(multi_input_template.format(**case))
```

## Task 3: Prompt Variations Engine

```python
# Task 3: Multiple prompt styles

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as an engaging story"
)

topic = "Machine Learning"

print("Task 3 Outputs:\n")
print("Teaching:\n", teaching_template.format(topic=topic))
print("\nInterview:\n", interview_template.format(topic=topic))
print("\nStorytelling:\n", story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

```python
# Task 4: Chat-based prompting with roles

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role} who explains things clearly."),
    ("user", "Explain {topic}")
])

# Test
role = "teacher"
topic = "Neural Networks"

messages = chat_template.format_messages(role=role, topic=topic)

print("Task 4 Output:\n")
for msg in messages:
    print(msg)
```

## Task 5: Input Validation Layer

```python
# Task 5: Input validation

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'casual'")
        tone = "casual"

    return audience, tone

# Test
audience, tone = validate_inputs("kids", "funny")
print("Validated:", audience, tone)
```

## Task 6: Prompt Generator App

```python
# Task 6: Prompt generator function

# Templates dictionary (modular design)
templates = {
    "teaching": "Explain {topic} clearly step by step for {audience} in a {tone} tone",
    "interview": "Ask 3 interview questions about {topic} for {audience} in a {tone} tone",
    "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
}

def generate_prompt(topic, audience, tone, style):
    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Select template
    selected_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return selected_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test
print("Task 6 Output:\n")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

```python
# Task 7: Reusability test using ONE template

reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    {"topic": "AI", "audience": "beginner", "tone": "formal"},
    {"topic": "Blockchain", "audience": "expert", "tone": "technical"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Data Science", "audience": "intermediate", "tone": "casual"},
    {"topic": "Cybersecurity", "audience": "beginner", "tone": "formal"},
]

print("Task 7 Outputs:\n")
for inp in inputs:
    print(reusable_template.format(**inp))
```
