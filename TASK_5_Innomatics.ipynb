##  1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Sample Input
likes = [200, 300, 250, 300]

# Function Call
total, result = analyze_post_engagement(likes)

# Output
print("Total Likes:", total)
print("Post Status:", result)
```

## 2: Healthcare – Medicine Stock Alert System


```python
def medicine_stock_alert(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    return stock, status


# Sample Input
stock = 6

# Function Call
stock_value, result = medicine_stock_alert(stock)

print("Medicine Stock:", stock_value)
print("Status:", result)
```

## 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_list, required_level):
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Sample Input
rainfall = [70, 75, 68, 75]
required = 70

avg, result = rainfall_checker(rainfall, required)

print("Average Rainfall:", avg)
print("Rainfall Status:", result)
```

## 4: Social Media – Duplicate Account Detection


```python
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"


# Sample Input
users = ["john", "emma", "john", "alex"]

result = detect_duplicates(users)

print("Duplicate Accounts Found:", result)
```

## 5: Healthcare – Appointment Eligibility Checker

```python
def appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return age, status


# Sample Input
age = 21

age_value, result = appointment_eligibility(age)

print("Patient Age:", age_value)
print("Eligibility Status:", result)
```

##  6: Agriculture – Premium Crop Price Filter


```python
def premium_crop_filter(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


# Sample Input
crop_prices = [1500, 2500, 1800, 3200]

result = premium_crop_filter(crop_prices)

print("Premium Crops:", result)
```

## 7: System Monitoring – Application Health Checker


```python
def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status


# Sample Input
errors = 7

error_value, result = application_health_checker(errors)

print("Error Count:", error_value)
print("System Status:", result)
```

## 8: Banking – Daily Transaction Limit Checker

```python
def transaction_limit_checker(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


# Sample Input
transaction_amount = 60000

amount, result = transaction_limit_checker(transaction_amount)

print("Transaction Amount:", amount)
print("Transaction Status:", result)
```

## 9: E-Learning – Student Attendance Eligibility System


```python
def attendance_checker(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status


# Sample Input
attendance = ["P", "P", "A", "P", "P"]

percentage, result = attendance_checker(attendance)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", result)
```

## 10: Smart Electricity Bill Analyzer


```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Sample Input
units_consumed = 250

bill_amount, usage_status = electricity_bill(units_consumed)

print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
