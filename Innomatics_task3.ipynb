## 1: Employee Performance Bonus Eligibility


```python
# Taking dictionary input from user
n = int(input("Enter number of employees: "))
employees = {}

for i in range(n):
    name = input("Enter employee name: ")
    score = int(input("Enter performance score: "))
    employees[name] = score

# Step 1: Find highest score manually
names = list(employees.keys())
highest_score = employees[names[0]]

for name in employees:
    if employees[name] > highest_score:
        highest_score = employees[name]

# Step 2: Find all employees with highest score (handle ties)
top_performers = []

for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

# Step 3: Print in exact expected format
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Taking input from user
query = input("Enter search query: ")

# Convert to lowercase
query = query.lower()

# Remove common punctuation manually
for ch in ",.!?;:":
    query = query.replace(ch, "")

# Split into words
words = query.split()

frequency = {}

# Count frequency of each word
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Store only words searched more than once
repeated_keywords = {}

for word in frequency:
    if frequency[word] > 1:
        repeated_keywords[word] = frequency[word]

# Print result
print(repeated_keywords)
```

## Sensor Data Validation


```python
# Taking input from user
sensor_readings = input("Enter sensor readings separated by comma: ")

# Convert input into list of integers
sensor_readings = sensor_readings.replace(",", " ").split()
sensor_readings = [int(value) for value in sensor_readings]

valid_readings = []

# Loop through list using index
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:   # Check if even
        valid_readings.append((i, sensor_readings[i]))

# Print in exact expected format
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## 4: Email Domain Usage Analysis


```python
# Taking input from user
n = int(input("Enter number of email IDs: "))
emails = []

for i in range(n):
    email = input("Enter email ID: ")
    emails.append(email)

domain_count = {}

# Extract domain and count
for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate percentage and print in expected format
for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

## 5: Sales Spike Detection


```python
# Taking input from user
sales = input("Enter daily sales separated by comma: ")

# Convert input into list of integers
sales = sales.replace(",", " ").split()
sales = [int(value) for value in sales]

# Step 1: Calculate average manually
total = 0

for value in sales:
    total += value

average = total / len(sales)

# Step 2: Detect days with sales more than 30% above average
threshold = average + (0.30 * average)

for i in range(len(sales)):
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```

## 6: Duplicate User ID Detection


```python
# Taking input from user
user_ids = input("Enter user IDs separated by comma: ")

# Convert input into list
user_ids = user_ids.replace(",", " ").split()

frequency = {}

# Count occurrences
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Display duplicates
for user in frequency:
    if frequency[user] > 1:
        print(f"{user} → {frequency[user]} times")

```
