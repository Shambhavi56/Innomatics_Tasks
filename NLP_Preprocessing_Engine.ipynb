4. Difference between Stemming and Lemmatization
---
* **Stemming** reduces words by cutting off endings, often producing incomplete or non-real words.
Example: “running” → “run”, but sometimes “better” → “bett”

* **Lemmatization** converts words into their base or dictionary form using linguistic rules.
Example: “running” → “run”, “better” → “good”
 Lemmatization is more accurate and meaningful than stemming.



# **NLP Preprocessing Engine**

## Objective
Build a robust and scalable NLP preprocessing pipeline to clean noisy real-world text.

## Key Features
- Noise removal (URLs, numbers, emojis)
- Text normalization
- Tokenization + Lemmatization (spaCy)
- Token analytics
- Frequency analysis

##1. Install & Import Dependencies

```python
import re
import spacy
from collections import Counter
```

### We use spaCy for efficient tokenization and lemmatization.

```python
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

## Preprocessing  Function

```python
def preprocess_text(text):
    """
    Cleans and processes raw text into tokens and normalized sentence.
    Handles noise like URLs, emojis, repeated characters, etc.
    """

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Normalize repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters
    text = re.sub(r"[^a-z\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # spaCy processing
    doc = nlp(text)

    tokens = []
    for token in doc:
        word = token.lemma_

        if len(word) <= 2 and word not in {"no", "not"}:
            continue

        tokens.append(word)

    return tokens, " ".join(tokens)
```

## Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
```

## **Token Analytics**

```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)
```

## Automated Analysis

```python
def noise_score(original, cleaned):
    return len(original) - len(cleaned)

def retention_ratio(original, tokens):
    return len(tokens) / max(len(original.split()), 1)

analysis_results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    analysis_results.append({
        "original": sentence,
        "clean": clean,
        "tokens": tokens,
        "noise": noise_score(sentence, clean),
        "token_count": len(tokens),
        "retention": retention_ratio(sentence, tokens)
    })
```

```python
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    total, unique, avg_len = token_analytics(tokens)

    print("\nSentence:", sentence)
    print("Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg_len)
```

```python
most_noisy = max(analysis_results, key=lambda x: x["noise"])
most_meaningful = max(analysis_results, key=lambda x: (x["token_count"], x["retention"]))

print("Most Noisy:", most_noisy)
print("\nMost Meaningful:", most_meaningful)
```

## **Observations**

- The most noisy sentence is:
  'Visit https://openai.com now!'

- The most meaningful sentence is:
  'Get 100% FREE access now!!!'

These observations are based on noise score, token count, and retention ratio after preprocessing.

## **Frequency Analysis**

```python
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Least 5 Frequent Words:", counter.most_common()[-5:])
```

## Full Pipeline


```python
def full_pipeline(text_list):
    tokens_all = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        tokens_all.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": tokens_all,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("\nAll Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for sentence in result["clean_sentences"]:
    print("-", sentence)
```

## Error Handling

```python
def process_batch(text_list):
    results = []

    for idx, text in enumerate(text_list):
        try:
            tokens, clean = preprocess_text(text)

            results.append({
                "index": idx,
                "original": text,
                "tokens": tokens,
                "clean_text": clean
            })

        except Exception as e:
            results.append({
                "index": idx,
                "original": text,
                "tokens": [],
                "clean_text": "",
                "error": str(e)
            })

    return results
```

## Edge Cases

```python
edge_cases = ["", None, "😍😍😍", 12345]

results = process_batch(edge_cases)

for r in results:
    print(r)
```
