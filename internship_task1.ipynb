
# task 1.1 User-Login-Check



```python

class AdminLogin:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def login(self, user_username, user_password):
        return user_username == self.username and user_password == self.password


# Create object
admin = AdminLogin("admin","1234")

# Taking input
username = input("Enter username: ")
password = input("Enter password: ")


# Call login method
if admin.login(username, password):
  print("Login Successful")
else:
  print("invalid Credentials")
```

# task 1.2 Pass/Fail Analyzer

```python
def pass_fail_analyser(marks_list: list) -> None:
    pass_count = 0
    fail_count = 0

    for marks in marks_list:
        if marks > 50:
            pass_count += 1
        else:
            fail_count += 1

    print(f"Passing count: {pass_count}")
    print(f"Failing count: {fail_count}")


number_of_students = int(input("Enter number of students appeared in exam: "))

marks = []

for ele in range(1, number_of_students + 1):
    mark = int(input(f"Enter marks of student {ele}: "))
    marks.append(mark)

pass_fail_analyser(marks)



```

# Task 1.3 Simple Data Cleaner

```python
def remove_spaces(name: str) -> str:
    return name.strip()


def to_lowercase(name: str) -> str:
    return name.lower()


def standardize_name(name: str) -> str:
    name = remove_spaces(name)
    name = to_lowercase(name)
    return name


def clean_usernames(names_list: list) -> list:
    cleaned_list = []

    for name in names_list:
        cleaned_name = standardize_name(name)
        cleaned_list.append(cleaned_name)

    return cleaned_list


# Input Data
names = [" Alice ", "bob", " CHARLIE "]

# Call Parent Function
cleaned_names = clean_usernames(names)

print("Cleaned Names:", cleaned_names)
```

# Task 1.4 Message Length Analyzer

```python
def message_length_analyzer(messages: list) -> None:
    for message in messages:
        length = len(message)
        print(f"Message: '{message}' , Length: {length}")

        if length > 10:
            print(" This message is longer than 10 characters.")
        print()  # for spacing


messages = ["Hi", "Welcome to the platform", "OK"]

message_length_analyzer(messages)
```

# Task 1.5 Error Message Detector


```python
def error_message_detector(logs: list) -> int:
    error_count = logs.count("ERROR")
    print(f"Number of ERROR entries: {error_count}")
    return error_count


logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_message_detector(logs)
```
